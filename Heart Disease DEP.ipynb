```python
Sex = pd.get_dummies(df_heart['sex'],drop_first=True)  
Sex
```

```python
cp1 = pd.get_dummies(df_heart['cp'],drop_first=True)
df_heart.drop(['sex','trestbps','fbs','restecg','thalach','oldpeak','slope','ca','cp','thal'],axis=1,inplace=True) 
df_heart1 = pd.concat([df_heart,Sex,cp1],axis=1)
df_heart1
```

```python
 
X = df_heart1.drop('target',axis=1)  
y = df_heart1['target'] 
```

```python
 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101) 
```

```python
 

from sklearn.linear_model import LogisticRegression 
```

```python
lm = LogisticRegression() 
```

```python
 
lm.fit(X_train,y_train)  
```

```python
 
predict = lm.predict(X_test)
 
from sklearn.metrics import classification_report
print(classification_report(y_test,predict)) 
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predict))
 
```

```python
 
X = df_heart1.drop('thalach',axis=1) 
y = df_heart1['thalach']  
X
```

```python
 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()  
lm.fit(X_train,y_train)  
```

```python
 
print(lm.intercept_)  
```

```python
print(lm.coef_)
```

```python
 
predictions = lm.predict( X_test)  
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient']) 
coeff_df                                                             
```
